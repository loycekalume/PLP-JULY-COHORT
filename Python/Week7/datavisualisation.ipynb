```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Load the iris dataset from sklearn
iris = load_iris(as_frame=True)
df = iris.frame 
```

```python
# Display first few rows
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

```python
# Check dataset info
print("\nDataset Info:")
print(df.info())
```

```python

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Compute basic statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
grouped = df.groupby("target").mean()
print("\nMean values per species:")
print(grouped)
```

```python
# Map numeric target back to species names
df["species"] = df["target"].map({i: name for i, name in enumerate(iris.target_names)})

print("\nInteresting Finding:")
print("We can see clear differences in petal size across species, "
      "which is why Iris dataset is useful for classification tasks.")
```

```python
# 1. Line chart (trend over index, not time here)
plt.figure(figsize=(8,5))
df["sepal length (cm)"].plot(kind="line", title="Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python

# 2. Bar chart - average petal length per species
plt.figure(figsize=(8,5))
df.groupby("species")["petal length (cm)"].mean().plot(kind="bar", color=["skyblue","orange","green"])
plt.title("Average Petal Length per Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
# 3. Histogram of sepal width
plt.figure(figsize=(8,5))
df["sepal width (cm)"].plot(kind="hist", bins=20, color="purple", alpha=0.7)
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.show()
```

```python
# 4. Scatter plot: sepal length vs petal length
plt.figure(figsize=(8,5))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=df)
plt.title("Sepal Length vs Petal Length by Species")
plt.show()
```
